```python
import pandas as pd
import requests
```

```python
data = pd.read_csv('../clean_data/velo_df.csv')
```

```python
data_filtered = data.loc[data['dep'].isin([750])]
```

```python
data_filtered_lat = data_filtered.loc[data_filtered['lat'].isna()]
```

```python
data_filtered_latlong = data_filtered_lat.loc[data_filtered_lat['long'].isna()]
```

### Call API BAN lat, lon manquantes

```python
data_filtered_latlong.adr.value_counts() == 51
```

```python
data_filtered.loc[data_filtered['adr'] == 'ROUTE DE SEVRES A NEUILL']
```

```python
def get_lat_lon(address, url='https://api-adresse.data.gouv.fr/search'):

    params = {'q': address, 'citycode': 75056}
    r = requests.get(url, params)
    try: return r.json()['features'][0]['geometry']['coordinates'][::-1]
    except: print(f'ERROR : {address}, STATUS CODE : {r.status_code}')

get_lat_lon('22 r e pasy')
```

```python
data_filtered_latlong['latlong'] = data_filtered_latlong['adr'].map(lambda address: get_lat_lon(address))
```

```python
data_filtered_latlong.head()
```

```python
data_filtered_latlong.dtypes
```

```python
data_filtered_latlong['latlong'] = data_filtered_latlong['latlong'].astype(str)

data_filtered_latlong['latlong'] = data_filtered_latlong['latlong'].str.strip('[]').str.replace(' ', '')

data_filtered_latlong[['lat', 'long']] = data_filtered_latlong['latlong'].str.split(',', expand=True)
```

```python
data_filtered_latlong = data_filtered_latlong.drop(columns=['latlong'], inplace=False)
```

```python
data_filtered_latlong = data_filtered_latlong.drop(columns=['latitude'], inplace=False)
```

```python
data_filtered_latlong.dropna(subset=['lat', 'long'], inplace=True)
```
